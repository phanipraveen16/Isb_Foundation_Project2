```python
df = yf.download("RELIANCE.NS RCOM.NS HDFCBANK.NS INFY.NS NESTLEIND.NS TATAMOTORS.NS HINDUNILVR.NS", start="2021-01-01", end="2021-08-27",group_by="ticker")
```

```python
hdfc = pd.DataFrame(df['HDFCBANK.NS']['Close']).reset_index()
reliance = pd.DataFrame(df['RELIANCE.NS']['Close']).reset_index()
tatamotors = pd.DataFrame(df['TATAMOTORS.NS']['Close']).reset_index()
rel_com = pd.DataFrame(df['RCOM.NS']['Close']).reset_index()
infy = pd.DataFrame(df['INFY.NS']['Close']).reset_index()
nestle = pd.DataFrame(df['NESTLEIND.NS']['Close']).reset_index()
hul = pd.DataFrame(df['HINDUNILVR.NS']['Close']).reset_index()
```

```python
fig, ax = plt.subplots()
ax.plot_date(hdfc['Date'], hdfc['Close'], marker='', linestyle='-')
avg = hdfc['Close'].mean()
ax.plot(hdfc['Date'], [avg]*len(hdfc), color='red', lw=2, ls='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('HDFC Share Price Trend - 2021')
plt.show()

fig, ax = plt.subplots()
ax.plot_date(reliance['Date'], reliance['Close'], marker='', linestyle='-')
avg = reliance['Close'].mean()
ax.plot(reliance['Date'], [avg]*len(reliance), color='red', lw=2, ls='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('Reliance Share Price Trend - 2021')
plt.show()

fig, ax = plt.subplots()
ax.plot_date(tatamotors['Date'], tatamotors['Close'], marker='', linestyle='-')
avg = tatamotors['Close'].mean()
ax.plot(tatamotors['Date'], [avg]*len(tatamotors), color='red', lw=2, ls='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('Tatamotors Share Price Trend - 2021')
plt.show()

fig, ax = plt.subplots()
ax.plot_date(rel_com['Date'], rel_com['Close'], marker='', linestyle='-')
avg = rel_com['Close'].mean()
ax.plot(rel_com['Date'], [avg]*len(rel_com), color='red', lw=2, ls='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('Reliance Communications Share Price Trend - 2021')
plt.show()

fig, ax = plt.subplots()
ax.plot_date(infy['Date'], infy['Close'], marker='', linestyle='-')
avg = infy['Close'].mean()
ax.plot(infy['Date'], [avg]*len(infy), color='red', lw=2, ls='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('Infy Share Price Trend - 2021')
plt.show()


fig, ax = plt.subplots()
ax.plot_date(nestle['Date'], nestle['Close'], marker='', linestyle='-')
avg = nestle['Close'].mean()
ax.plot(nestle['Date'], [avg]*len(nestle), color='red', lw=2, linestyle='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('Nestle Share Price Trend - 2021')
plt.show()

fig, ax = plt.subplots()
ax.plot_date(hul['Date'], hul['Close'], marker='', linestyle='-')
avg = hul['Close'].mean()
ax.plot(hul['Date'], [avg]*len(hul), color='red', lw=2, linestyle='--', label="average plot")
fig.autofmt_xdate()
plt.ylabel('Closing Price')
plt.title('Hul Share Price Trend - 2021')
plt.show()
```

```python
# Correlation between Infy and Nestle

plt.scatter(infy['Close'], nestle['Close'])
plt.show()

from scipy.stats import pearsonr
corr, _ = pearsonr(infy['Close'], nestle['Close'])
print('Pearsons correlation: %.3f' % corr)
```

```python
# Correlation between HUL and Nestle

plt.scatter(hul['Close'], nestle['Close'])
plt.show()

from scipy.stats import pearsonr
corr, _ = pearsonr(hul['Close'], nestle['Close'])
print('Pearsons correlation: %.3f' % corr)
```
